```python
A = torch.tensor([1,2,3,4,5])
B = torch.tensor([10,20,30])

C, D = torch.meshgrid(A, B, indexing='ij')
print(C)
print(D)
```

```python
C_s = C.flatten()
D_s = D.flatten()

print(C_s)
print(D_s)
```

```python
shifts = torch.stack((C_s, D_s, C_s, D_s), dim=1)
print(shifts)
```
